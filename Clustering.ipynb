```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Verzeo July ML 2/Data/k means 1.csv')
```

```python
df.head()
```

```python
plt.scatter(df['X'], df['Y'])
plt.show
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2)
```

```python
km.fit(df)
```

```python
y_pred = km.predict(df)
```

```python
y_pred
```

```python
df['cluster'] = y_pred
y_pred
```

```python
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
```

```python
df1
```

```python
df2
```

```python
plt.scatter(df1['X'], df1['Y'], color = "g")
plt.scatter(df2['X'], df2['Y'], color = 'r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```

```python
c1 = km.cluster_centers_[:, 0]
c2 = km.cluster_centers_[:, 1]
```

```python
plt.scatter(df1['X'], df1['Y'], color = "g")
plt.scatter(df2['X'], df2['Y'], color = 'r')
plt.scatter(c1,c2,color = "violet", marker = "*")
plt.xlabel("X")
plt.ylabel("Y")
plt.show
```

```python
wcss=[]
for i in range(1,4):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,4), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```
