# Seaborn with python

```python
cars= sns.load_dataset('mpg')
```

```python
type(cars)
```

```python
cars.dropna(inplace = True)
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.relplot(x = 'model_year', y ='mpg', hue = 'cylinders', col = 'origin', data = cars)
```

```python
sns.set_style('whitegrid')

cyl_4 = cars[cars.cylinders==4]
cyl_8 = cars[cars.cylinders==8]

plt.figure(figsize=(8,6))

sns.kdeplot(cyl_4.horsepower, cyl_4.mpg,
            cmap="Blues", shade=True, shade_lowest=False)
sns.kdeplot(cyl_8.horsepower, cyl_8.mpg,
            cmap="Reds", shade=True, shade_lowest=False)

plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')

plt.annotate("4 Cylinders", (105, 32), color='b', fontsize=16, fontweight='bold')
plt.annotate("8 Cylinders", (190, 18), color='r', fontsize=16, fontweight='bold')
plt.tight_layout()
```

```python
sns.kdeplot(cars.horsepower, shade = True, bw=10, cumulative=True)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20, shade=True)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20, shade=True, shade_lowest=False)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20, shade=True, shade_lowest=False, cbar=True)
```

```python
sns.set_style('whitegrid')
```

```python
cyl4 = cars[cars.cylinders == 4]
cyl8 = cars[cars.cylinders == 8]
plt.figure(figsize=(8,6))
sns.kdeplot(cyl4.horsepower, cyl4.mpg, cmap ="Blues", shade= True, shade_lowest=False)
sns.kdeplot(cyl8.horsepower, cyl8.mpg, cmap ="Reds", shade= True, shade_lowest=False)
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon')
plt.annotate("4 cylinders", (100,40), color="b", fontsize= 16, fontweight = "bold")
plt.annotate("8 cylinders", (200,25), color="r", fontsize= 16, fontweight = "bold")
```

```python
sns.pairplot(cars[cars.cylinders.isin([4,6,8])], 
             hue='cylinders', 
             vars=['mpg', 'horsepower', 'acceleration'], 
             palette='inferno_r'
            );
```

```python
usa = cars[cars.origin == 'usa']

sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width', inner=None, color='xkcd:forest green'
              )
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               scale='width', inner=None, color='xkcd:forest green'
              )
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='white', size=3
             )
```
